# Pure, Mixed, and Entangled States

```python
from qiskit import Aer, BasicAer, execute, QuantumCircuit, transpile
import qiskit.quantum_info as qi
import qiskit.tools
from qiskit.visualization import plot_bloch_multivector, plot_bloch_vector, plot_histogram
from math import sqrt
from numpy import log, log2
import numpy as np
import matplotlib.pyplot as plt
```

## Pure States

The quantum analogs to classical bits set to 0 and 1 are the |0> and |1> states, respectively.

```python
psi_zero = qi.Statevector.from_label('0')
plot_bloch_multivector(psi_zero)
```

```python
psi_one = qi.Statevector.from_label('1')
plot_bloch_multivector(psi_one)
```

### Measurements

One way to think about measurment is in the projection of the state vector along the z-axis.  A qubit initialized to the |0> state will always be measured as a 0, assuming proper measurement technique and lack of technical malfunction, just as a qubit initialized to the |1> state will always yield a 1.  Note that Qiskit performs 1,024 measurements by default.

```python
# Create a quantum circuit with one qubit initialized to the |0> state
# and one classical bit in which to record the measurement.
# Note how 100% of the measurements yielded a '0'

qc = QuantumCircuit(1, 1)
qc.measure(0, 0)
backend = BasicAer.get_backend('qasm_simulator')
tqc = transpile(qc, backend)
counts = backend.run(tqc).result().get_counts()
plot_histogram(counts)
```

```python
# Create a quantum circuit with one qubit initialized to the |1> state,
# which is achieved by applying a Pauli X operator to the initialized |0> state,
# and one classical bit in which to record the measurement.
# Note how 100% of the measurements yielded a '1'

qc = QuantumCircuit(1, 1)
qc.x(0)
qc.measure(0, 0)
backend = BasicAer.get_backend('qasm_simulator')
tqc = transpile(qc, backend)
counts = backend.run(tqc).result().get_counts()
plot_histogram(counts)
```

### Superposition

If a vector pointing to the north pole of a qubit always yields a '0' and one pointing to the south pole always yields a '1', then one pointing to a spot on the equator should yield a '0' ~50% of the time and a '1' ~50% of the time.  This is achieved by rotating the qubit in the |0> state by 90 degrees to the equator with a **Hadamard operator**.

```python
qc = QuantumCircuit(1, 1)
qc.h(0)
plot_bloch_multivector(qi.Statevector(qc))
```

```python
# Measure the quantum state in a perfect superposition of |0> and |1>.
# Notice how '0's and '1's are measured ~50% of the time
qc.measure(0, 0)
backend = BasicAer.get_backend('qasm_simulator')
tqc = transpile(qc, backend)
counts = backend.run(tqc).result().get_counts()
plot_histogram(counts)
```

### Density Matrices

Density matrices are useful in determining the purity of a quantum system.  Further, a mixed state cannot be expressed as a state vector and must be expressed with a density matrix.

The density matrix for the |+> state is created below.

```python
qc = QuantumCircuit(1, 1)
qc.h(0)
dm_plus = qi.DensityMatrix(qi.Statevector(qc))
dm_plus
```

Von Neumann entropy takes the density matrix as an argument and returns either a "0", indicating a pure state, or a non-zero value less than or equal to "1" for a mixed state.

The entropy of the |+> state is calculated below.  As this is a pure state, the von Neumann entropy is "0"; however, floating point roundoff error yields a non-zero, albeit trivial, value.

```python
entropy = qi.entropy(dm_plus)
entropy
```

## Mixed States

Unlike a pure state which is entirely known and may be represented with a state vector, a mixed state is an ensemble of pure states with a probablity 
associated to each one.  Mixed states represent uncertainty in a quantum system, like when noise is introduced.

To exemplify this, let's calculate the entropy of a mixed state where there is a 30% chance the system is in the |0> pure state and 70% chance it is in the |1> pure state.

```python
prob_0 = 0.3
prob_1 = 0.7

mixed_state = prob_0 * qi.DensityMatrix.from_label('0') + prob_1 * qi.DensityMatrix.from_label('1')
entropy = qi.entropy(mixed_state)
entropy
```

Let's vary these probabilities that the system is in the |0> state and plot their entropies.

```python
def entropy_mixed_state(prob_0, prob_1):
    mixed_state = prob_0 * qi.DensityMatrix.from_label('0') + prob_1 * qi.DensityMatrix.from_label('1')
    return qi.entropy(mixed_state)

probs_0 = np.linspace(0, 1, 101)
entropies = [entropy_mixed_state(p0, p1) for p0, p1 in zip(probs_0, 1-probs_0)]

plt.plot(probs_0, entropies)
plt.xlabel('Probability of |0> State')
plt.ylabel('von Neumann Entropy')
plt.title('von Neumann Entropy vs Probability of |0> State')
plt.grid(True)
plt.show()
```

## Entanglement

Entanglement refers to a collection of qubits that cannot be described without one another.  Mathematically, this is shown by the inability to factor the density matrix of an entangled state into constituent product states.

The maximally entangled **|psi_+> Bell state** is used as an example.

```python
# Observe that Qiskit uses little endian notation, meaning the order of the qubits is reversed
psi_plus = 1/sqrt(2)*(qi.Statevector.from_label('10') + qi.Statevector.from_label('01'))
density_matrix_psi_plus = qi.DensityMatrix(psi_plus)
density_matrix_psi_plus
```

```python
entropy = qi.entropy(density_matrix_psi_plus)
entropy
```

Note that entropy of this entangled Bell state is 0, indicating it is a pure state.  What happens when calculating the entropy of qubit A by "tracing out" qubit B?  To answer this question, the **partial trace** is used.

```python
density_matrix_a = qi.partial_trace(density_matrix_psi_plus, [0])
density_matrix_a
```

Solving for entropy of qubit A:

```python
entropy = qi.entropy(density_matrix_a)
entropy
```

The results above are consistent with the definition of entanglement where neither of the states of the entangled particles can be interpreted without the other(s).  A traced out qubit is just one example of a mixed state where uncertainty is represented via a density matrix.

## Conclusions

Pure states, including entangled states, have entropies of 0, whereas mixed states have non-zero entropies.  The partial trace allows us to "trace out" a qubit and measure entropy on the remaining qubit(s) to determine purity.
